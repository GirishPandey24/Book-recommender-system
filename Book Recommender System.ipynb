```python
import requests
from bs4 import BeautifulSoup
import json
import time
```

```python
page=requests.get("https://www.goodreads.com/list/show/13086.Goodreads_Top_100_Literary_Novels_of_All_Time")
page## 200 is the code that means we are successfully access the page data
```

```python
page.content
```

```python
#creating object of beautiful soup
soup=BeautifulSoup(page.content,'html.parser')#'html.parser' bcoz the data in html form....write xml.parser if it is in xml
print(soup.prettify())#prettify() is used to print the data in well formated way 
```

```python
booklist=soup.find_all('a',class_='bookTitle')
#or
# booklist=soup.find_all('a',{'class':'bookTitle'})# fetch the data by using dictionary 
booklist
```

```python
with open('all_books.txt','w') as f:
    for book in booklist:
        f.write(book['href']+'\n')
```

```python
def get_book_data():
    with open('all_books.txt') as f:
        booklist=f.readlines()#readlines() function will read the line step by step and store it into the list[]

    all_data=[]
    for idx,book in enumerate(booklist):
        print(f"Extracting data for books {idx+1}")
        r=requests.get("https://www.goodreads.com/"+book)
        soup=BeautifulSoup(r.text,'html.parser')
#         time.sleep(7)
        data={
            'title':soup.find('h1',{'id':'bookTitle'}).get_text().strip(),
            'author':soup.find('span',{'itemprop':'name'}).get_text().strip(),
            'rating':soup.find('span',{'itemprop':'ratingValue'}).get_text().strip(),
            'genre':soup.find('a',{'class':'actionLinkLite bookPageGenreLink'}).get_text().strip()
        }
        all_data.append(data)
        with open('result.json','w') as f:
            f.write((json.dumps(all_data)))
    
get_book_data()
```

```python
import pandas as pd
df=pd.read_json("result.json")
```

```python
def recommender(genre):
    '''
    recommend top rated book from same genre
    '''
    top= df[df['genre']==genre].sort_values(by='rating',ascending=False)
    if top.shape[0] > 1:
        return dict(top.iloc[0])
    else:
        return dict(df.sort_values(by='rating',ascending=False).iloc[:10].sample(1).squeeze())
    
```

```python
recommender('Classics')
```

```python
recommender('Academic')
```
